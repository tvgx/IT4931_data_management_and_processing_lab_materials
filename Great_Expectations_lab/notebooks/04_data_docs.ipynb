```python
import subprocess
import os
from pathlib import Path

project_root = Path('.').resolve()
if project_root.name != 'Great_Expectations_lab':
    project_root = project_root.parent / 'Great_Expectations_lab'

print("üìö Data Docs Overview:")
print("=" * 60)
print("""
Data Docs l√† HTML documentation t·ª± ƒë·ªông generate t·ª´:
- Expectations
- Validation results
- Data profiling

Features:
- Interactive documentation
- Validation history
- Data quality metrics
- Shareable v·ªõi team
""")

print("\nüí° Commands:")
print("  - great_expectations docs build: Generate docs")
print("  - great_expectations docs serve: Serve docs locally")
print("  - Docs location: great_expectations/uncommitted/data_docs/")

os.chdir(project_root)

# Example: Generate docs
def generate_docs():
    """Generate GE documentation"""
    try:
        result = subprocess.run(
            ['great_expectations', 'docs', 'build'],
            capture_output=True,
            text=True,
            timeout=60
        )
        
        if result.returncode == 0:
            print("\n‚úÖ Documentation generated successfully!")
            print("üí° Serve v·ªõi: great_expectations docs serve")
            print("   Then open http://localhost:8000")
        else:
            print("\n‚ö†Ô∏è  Documentation generation failed")
            print(result.stderr)
    except Exception as e:
        print(f"‚ö†Ô∏è  Error: {e}")

# Uncomment ƒë·ªÉ generate:
# generate_docs()
```
