## 1. Kiá»ƒm tra Installation


````python
# Kiá»ƒm tra dbt version
import subprocess
import sys

def check_dbt_installation():
    """Kiá»ƒm tra dbt Ä‘Ã£ Ä‘Æ°á»£c cÃ i Ä‘áº·t chÆ°a"""
    try:
        result = subprocess.run(
            ['dbt', '--version'],
            capture_output=True,
            text=True,
            check=True
        )
        print("âœ… dbt Ä‘Ã£ Ä‘Æ°á»£c cÃ i Ä‘áº·t:")
        print(result.stdout)
        return True
    except FileNotFoundError:
        print("âŒ dbt chÆ°a Ä‘Æ°á»£c cÃ i Ä‘áº·t")
        print("ğŸ’¡ Cháº¡y: pip install dbt-core dbt-postgres")
        return False
    except subprocess.CalledProcessError as e:
        print(f"âš ï¸  Lá»—i khi kiá»ƒm tra dbt: {e}")
        return False

check_dbt_installation()
````

## 2. Test dbt Connection

Sá»­ dá»¥ng `dbt debug` Ä‘á»ƒ kiá»ƒm tra káº¿t ná»‘i database.


````python
# Test dbt connection
import os
from pathlib import Path

def test_dbt_connection():
    """Test dbt connection vá»›i database"""
    print("ğŸ§ª Testing dbt connection...")
    print("=" * 60)
    
    # Get project root
    project_root = Path('.').resolve()
    if project_root.name != 'dbt_lab':
        # Try parent directory
        project_root = project_root.parent / 'dbt_lab'
    
    # Change to project directory
    original_dir = os.getcwd()
    try:
        if project_root.exists():
            os.chdir(project_root)
        
        # Run dbt debug
        result = subprocess.run(
            ['dbt', 'debug', '--profiles-dir', '.', '--project-dir', '.'],
            capture_output=True,
            text=True,
            timeout=30
        )
        
        print(result.stdout)
        
        if result.returncode == 0:
            print("\nâœ… dbt connection successful!")
        else:
            print("\nâŒ dbt connection failed!")
            print("\nError output:")
            print(result.stderr)
            print("\nğŸ’¡ Make sure:")
            print("  1. Docker containers are running: docker-compose up -d")
            print("  2. PostgreSQL is accessible on port 5432")
            print("  3. profiles.yml has correct connection settings")
        
    except subprocess.TimeoutExpired:
        print("â±ï¸  Connection test timed out")
    except Exception as e:
        print(f"âš ï¸  Error: {e}")
    finally:
        os.chdir(original_dir)

test_dbt_connection()
````
