```python
from pyspark.sql import SparkSession

# Spark session vá»›i Iceberg support
spark = SparkSession.builder \
    .appName("LakehouseProcessing") \
    .config("spark.sql.extensions", "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.iceberg.spark.SparkSessionCatalog") \
    .getOrCreate()

print("âœ… Spark session created vá»›i Iceberg support")

# Example: Read tá»« Kafka vÃ  write to Iceberg
print("""
# Spark Structured Streaming tá»« Kafka
df = spark.readStream \
    .format("kafka") \
    .option("kafka.bootstrap.servers", "localhost:9092") \
    .option("subscribe", "events") \
    .load()

# Process vÃ  write to Iceberg
df.writeStream \
    .format("iceberg") \
    .outputMode("append") \
    .option("path", "lakehouse.bronze.events") \
    .start()
""")
```
