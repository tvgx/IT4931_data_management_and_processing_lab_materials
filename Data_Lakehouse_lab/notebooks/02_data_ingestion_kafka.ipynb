```python
from kafka import KafkaProducer
import json
import time

# Kafka producer example
producer = KafkaProducer(
    bootstrap_servers=['localhost:9092'],
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

# Sample data
events = [
    {'user_id': 1, 'event_type': 'page_view', 'timestamp': '2024-01-10T10:00:00'},
    {'user_id': 2, 'event_type': 'purchase', 'timestamp': '2024-01-10T10:05:00'},
]

print("ðŸ“¥ Ingesting data to Kafka...")
for event in events:
    producer.send('events', event)
    print(f"Sent: {event}")

producer.flush()
print("âœ… Data ingested successfully!")
```
